# Visualization: Earnings Based On College Majors

In this project, We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. 

Using visualizations, we want to explore questions from the dataset like:
- Do students in more popular majors make more money?
    - Using scatter plots
- How many majors are predominantly male? Predominantly female?
    - Using histograms
- Which category of majors have the most students?
    - Using bar plots

Explanations for the variable names what will be used later:

- Numerical:
    - **Rank** - Rank by median earnings (the dataset is ordered by this column).
    - **Total** - Total number of people with major.
    - **Sample_size** - Sample size (unweighted) of full-time.
    - **Men** - Male graduates.
    - **Women** - Female graduates.
    - **ShareWomen** - Women as share of total.
    - **Employed** - Number employed.
    - **Median** - Median salary of full-time, year-round workers.
    - **Low_wage_jobs** - Number in low-wage service jobs.
    - **Full_time** - Number employed 35 hours or more.
    - **Part_time** - Number employed less than 35 hours.
    
- Categorial:
    - **Major_code** - Major code.
    - **Major** - Major description.
    - **Major_category** - Category of major.

### Discover the relationship between two vairables using scatter plots

We want to use the scatter plots to explore the following questions:
- Do students in more popular majors make more money?
- Do students that majored in subjects that were majority female make more money?
- Is there any link between the number of full-time employees and median salary?

1.Do students in more popular majors make more money?

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter', 
                  title='Median vs. Sample size', figsize=(8,5))
ax.set_title('Median Salary vs. Major Popularity')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', 
                  title='Unemployment Rate vs. Major Popularity', figsize=(8,5))
```

no

2.Do students that majored in subjects that were majority female make more money?

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', 
                  title='Unemployment Rate vs. Women population', figsize=(8,5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', 
                  title='Median Salary vs. Male Graduate Population', figsize=(8,5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', 
                  title='Median Salary vs. Female Graduate Population', figsize=(8,5))
```
